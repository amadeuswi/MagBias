```python
#we interpolate the new galaxy bias which is max(bg_old, C*sg) with
#C = 1/2
Crat = 1/2


#save files to:

znam = LSST['bgnew_file_z']
magnam = LSST['bgnew_file_magmax']
bgnewnam = LSST['bgnew_file']


#make tables and mesh grid:


# ztab = np.linspace(0.0005,3.9, 1000) #full
# magtab = np.linspace(19,27,500)

ztab = np.linspace(0.0005,3.9, 100) #testing
magtab = np.linspace(21,27,100)

magmesh, zmesh = np.meshgrid(magtab, ztab)
```

```python




bgnewtab = np.zeros((len(ztab), len(magtab)))
sgtab = np.zeros((len(ztab), len(magtab)))
for iz in range(len(ztab)):
    print "working at redshift {}".format(ztab[iz])
    for im in range(len(magtab)):
        bgnewtab[iz,im] = np.float128(bgal_new(ztab[iz], magtab[im], CRATIO = Crat, experiment = LSST))
        sgtab[iz,im] = np.float128( sg(ztab[iz], experiment = LSST, MAXMAG = magtab[im]))
        if np.isinf(bgnewtab[iz,im]) or np.isnan(bgnewtab[iz,im]) or bgnewtab[iz,im]>10e50:
            print "for inf or nan"
            bgnewtab[iz,im] = 0
            print "settig bg to zero!"
```

```python
fig = P.figure(figsize = (9,9)); ax = fig.add_subplot(111, projection = '3d');

# ax.plot_wireframe(magmesh, zmesh, sgtab/bgnewtab, rstride=5, cstride=5, color = 'red')
ax.plot_surface(magmesh, zmesh, sgtab/bgnewtab, rstride=1, cstride=1, color = 'red')
ax.set_xlabel("m")
ax.set_ylabel("z")
ax.set_zlabel("sg/bgnew")

ax.view_init(15, 340)

# ax.set_zlim((0,30))
```

```python
np.savetxt(znam, ztab, header = "z table for bg interpolation")
np.savetxt(magnam, magtab, header = "magnitude cutoff table for bg interpolation")
np.savetxt(bgnewnam, bgnewtab, header = "bgal_new interpolation table")
```
